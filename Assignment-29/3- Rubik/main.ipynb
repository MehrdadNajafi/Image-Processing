```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_BGR = cv2.imread('rubix.png')
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
```

```python
rows, cols, _ = img_RGB.shape
```

```python
R, G, B = cv2.split(img_RGB)
```

```python
print('RED')
plt.imshow(R, cmap='gray')
plt.show()

print('GREEN')
plt.imshow(G, cmap='gray')
plt.show()

print('BLUE')
plt.imshow(B, cmap='gray')
plt.show()
```

```python
for i in range(rows):
    for j in range(cols):
        if R[i, j] == 255 and G[i, j] == 255 and B[i, j] == 0:
            img_RGB[i-2:i+3, j-2:j+3] = (0, 0, 255)
        
        elif R[i, j] == 255 and G[i, j] == 0 and B[i, j] == 255:
            img_RGB[i-2:i+3, j-2:j+3] = (0, 255, 0)
        
        elif R[i, j] == 0 and G[i, j] == 255 and B[i, j] == 255:
            img_RGB[i-2:i+3, j-2:j+3] = (255, 0, 0)
```
