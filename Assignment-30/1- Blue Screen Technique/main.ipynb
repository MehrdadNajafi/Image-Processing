```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_BGR = cv2.imread('SuperMan.jpg')
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
```

```python
background_img_BGR = cv2.imread('sky.jpg')
background_img_RGB = cv2.cvtColor(background_img_BGR, cv2.COLOR_BGR2RGB)

plt.imshow(background_img_RGB)
```

```python
rows, cols, _ = background_img_BGR.shape
img_BGR = cv2.resize(img_BGR, (cols, rows))

img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
img_HSV = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2HSV)

plt.imshow(img_RGB)
```

```python
H, S, V = cv2.split(img_HSV)

plt.imshow(H, cmap='gray')
plt.title('H')
plt.show()

plt.imshow(S, cmap='gray')
plt.title('S')
plt.show()

plt.imshow(V, cmap='gray')
plt.title('V')
plt.show()
```

```python
for i in range(rows):
    for j in range(cols):
        if 25 < H[i, j] < 75 and S[i, j] > 60 and V[i, j] > 80:
            img_RGB[i, j] = background_img_RGB[i, j]
            
plt.imshow(img_RGB)
```

```python
img_BGR = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2BGR)
cv2.imwrite('output.jpg', img_BGR)
```
